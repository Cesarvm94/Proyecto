## Análisis Estadístico
Se dispuso de un archivo csv con la recopilación de los tiempos de ejecución en cada base de datos propuesta y se realizaron los análisis estadísticos propuestos: ANOVA y El método de Tukey.

### Anova y Tukey

La prueba ANOVA es usado usualmente para comparar las medias de más de dos grupos de muestras. Se usará la prueba ANOVA univariada. El método consiste en la prueba de dos hipotesis, una nula y otra alternativa.

Hipótesis nula: Las medias de los grupos son iguales.  


<center> $H0: μ1=μ2=…=μp$ </center> 
  


Hipótesis alternativa: Al menos una media grupal es diferente de las otras.   

<center> $H1$: Todas las $μ$ no son iguales </center> 

Por otro lado, la prueba de Tukey busca determinar los grupos de muestras que se diferencian estadísticamente entre sí. Heredando las hipótesis usadas en la prueba anterior.

Estas dos pruebas tienen un rango de confianza expresado por la variable p. En ambas pruebas se tendrá la condición que si p <0,05 se rechaza la hipótesis nula y se indican diferencias estadísticamente significativas entre las poblaciones involucradas.

```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import statsmodels.api as sm
import statistics as stat
from statsmodels.formula.api import ols
ruta='C:/Users/cesar/Desktop/Tiempos/AnalisisPythonNube/BDIESD/SoloTiempos/1-1.txt'
tiempos=pd.read_csv(ruta, sep=",", header=None)

from scipy.stats import f_oneway
from statsmodels.stats.multicomp import pairwise_tukeyhsd
tiempos=tiempos.astype(float)
tiempos1=tiempos.transpose()
tiempos1.columns=["consulta1","consulta2","consulta3"]
tiempos_fixed=pd.melt(tiempos1.reset_index(), id_vars=['index'], value_vars=["consulta1","consulta2","consulta3"])
print(tiempos_fixed)
fvalue, pvalue = stats.f_oneway(tiempos1['consulta1'], tiempos1['consulta2'], tiempos1['consulta3'])
print('El valor f es: ',fvalue,'y el valor p es: ', pvalue)
print('Obteniendo un valor de p MAYOR a 0.05, aceptando la hipotesis nula en la que se afirma que \n las medias de las poblaciones son IGUALES')
print('La tabla ANOVA se presenta a continuación:')
model = ols('value ~ C(variable)', tiempos_fixed).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
print(anova_table)

print('\n')
print('La tabla del análisis de Tukey se presenta a continuación:')
tukey = pairwise_tukeyhsd(endog=tiempos_fixed['value'],
                          groups=tiempos_fixed['variable'],
                          alpha=0.05)
print(tukey)
print('Donde se evidencia que no se rechaza la hipótesis nula.')
y = [01.995,02.030,01.300,01.567,01.978,01.778,02.310,01.443,01.354,02.364,02.393,03.060,02.031,02.839,02.326,01.817,02.711,01.821,02.088,02.917]
z = [01.817,02.022,01.662,01.879,01.411,01.773,02.161,02.119,01.917,02.377,02.379,03.077,02.809,02.476,02.992,02.092,02.275,02.019,02.691,03.311]
m = [01.574,01.540,01.801,01.878,01.871,01.821,01.675,01.976,01.866,02.403,03.053,02.536,02.915,03.209,02.431,01.807,02.632,01.991,02.042,02.620]
print('\n')
print('Obtención de medias')
print('Obteniendo la media de la primer consulta: '+str(stat.mean(y)))
print('Obteniendo la media de la segunda consulta: '+str(stat.mean(z)))
print('Obteniendo la media de la tercer consulta: '+str(stat.mean(m)))
```
